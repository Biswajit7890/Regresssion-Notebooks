```python
## conducting Annova test with Target varibale (continuous) with Categorical variable
from scipy.stats import f_oneway
for i in cat_cols:
    print('*'*60)
    categoricalresult=zomato_new1.groupby(i)['Rating'].apply(list)
    Annovaresult=f_oneway(*categoricalresult)
    print("p value :", Annovaresult[1])
    if(Annovaresult[1]<0.05):
        print(i,"is correleated with the target variable")
    else:
        print(i,"is not correleated with the target variable")
    
```

```python
## Linear Rigression
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

Target='Rating'
Predictors=['Longitude','Latitude','count_of _items','Price range','Has Table booking','Has Online delivery']
X=zomato_new_ml[Predictors].values
y=zomato_new_ml[Target].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.21,random_state=7598)
mlr=LinearRegression()
predictmodel=mlr.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
print("The R squared :",metrics.r2_score(y_train,predictmodel.predict(X_train)))
print("Accuracy Score :",100-np.mean(abs(y_test-predictions)/y_test)*100)
X1=sm.add_constant(X_train)
Reg=sm.OLS(y_train,X1).fit() 
print("The Adjust R square ",Reg.rsquared_adj)
Trainingdataset=pd.DataFrame(data=zomato_new_ml,columns=X_test)



```

```python
## Decison tree

from sklearn.tree import DecisionTreeRegressor

Target='Rating'
Predictors=['Longitude','Latitude','count_of _items','Price range','Has Table booking','Has Online delivery']
X=zomato_new_ml[Predictors].values
y=zomato_new_ml[Target].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.35,random_state=75)
dcf=DecisionTreeRegressor(max_depth=4,criterion='mse')
predictmodel=dcf.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
print("R2 Value",metrics.r2_score(y_train,predictmodel.predict(X_train)))
print("The Accuracy score", 100-np.mean(np.abs(y_test-predictions)/y_test)*100)


Feature_importance=pd.Series(predictmodel.feature_importances_,index=Predictors)
Feature_importance.nlargest(10).plot(kind='bar')

                          
```

```python
## Graphical represntation of Decsion tree
from IPython.display import Image
from sklearn import tree
import pydotplus 

dot_data=tree.export_graphviz(dcf,feature_names=Predictors,class_names=Target)
Graph=pydotplus.graph_from_dot_data(dot_data)
Image(Graph.create_png())
```

```python
## random Forest regressor
from sklearn.ensemble import RandomForestRegressor

Target='Rating'
Predictors=['Longitude','Latitude','count_of _items','Price range','Has Table booking','Has Online delivery']
X=zomato_new_ml[Predictors].values
y=zomato_new_ml[Target].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.35,random_state=75)
rgf=RandomForestRegressor(n_estimators=150,max_depth=4,criterion='mse')
print(rgf)
predictmodel=rgf.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
print("R2 Value",metrics.r2_score(y_train,predictmodel.predict(X_train)))
print("The Accuracy score", 100-np.mean(np.abs(y_test-predictions)/y_test)*100)

Feature_importance=pd.Series(predictmodel.feature_importances_,index=Predictors)
Feature_importance.nlargest(10).plot(kind='bar')

```

```python
from IPython.display import Image
from sklearn import tree
import pydotplus

dot_data=tree.export_graphviz(rgf.estimators_[10],feature_names=Predictors, class_names=Target,out_file=None)
Graph=pydotplus.graph_from_dot_data(dot_data)
Image(Graph.create_png())
```

```python
## Ada Boost
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor

Target='Rating'
Predictors=['Longitude','Latitude','count_of _items','Price range','Has Table booking','Has Online delivery']
X=zomato_new_ml[Predictors].values
y=zomato_new_ml[Target].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=750000)
dcf=DecisionTreeRegressor( max_depth=3)
Ada=AdaBoostRegressor(base_estimator=dcf, n_estimators=150,learning_rate=0.02)
predictmodel=Ada.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
print("R2 Value",metrics.r2_score(y_train,predictmodel.predict(X_train)))
print("The Accuracy score", 100-np.mean(np.abs(y_test-predictions)/y_test)*100)

Feature_importance=pd.Series(predictmodel.feature_importances_,index=Predictors)
Feature_importance.nlargest(10).plot(kind='bar')
```

```python
## XgBoost Algorithm
from xgboost import XGBRegressor

Target='Rating'
Predictors=['Longitude','Latitude','count_of _items','Price range','Has Table booking','Has Online delivery']
X=zomato_new_ml[Predictors].values
y=zomato_new_ml[Target].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.45,random_state=750)
Xgb=XGBRegressor(max_depth=3, learning_rate=0.04, n_estimators=200, objective='reg:linear', booster='dart')
predictmodel=Xgb.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
print("R2 Value",metrics.r2_score(y_train,predictmodel.predict(X_train)))
print("The Accuracy score", 100-np.mean(np.abs(y_test-predictions)/y_test)*100)

Feature_importance=pd.Series(predictmodel.feature_importances_,index=Predictors)
Feature_importance.nlargest(10).plot(kind='bar')                

```

```python
##KNN

from sklearn.neighbors import KNeighborsRegressor

Target='Rating'
Predictors=['Longitude','Latitude','count_of _items','Price range','Has Table booking','Has Online delivery']
X=zomato_new_ml[Predictors].values
y=zomato_new_ml[Target].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=70000)
knn=KNeighborsRegressor(n_neighbors=6)
predictmodel=knn.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
print("The R2 score",metrics.r2_score(y_train,predictmodel.predict(X_train)))
print("The Accuracy score", 100-np.mean(np.abs(y_test-predictions)/y_test)*100)
```
