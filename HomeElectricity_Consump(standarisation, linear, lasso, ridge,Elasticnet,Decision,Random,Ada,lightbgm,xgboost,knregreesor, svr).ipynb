```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
```

```python
Home_Consumption_Ml=pd.read_csv('C:/Users/user/Desktop/IVY WORK BOOK/PYTHON/Python Datasets/Regression Datasets/HomeC.csv')
```

```python
Home_Consumption_Ml.dropna()
```

```python
Home_Consumption_Ml.isnull().sum()
```

```python
Del_cols=['time','use [kW]','precipProbability','precipProbability','cloudCover','windBearing','apparentTemperature','summary','icon','visibility','windBearing','precipIntensity']
Home_Consumption_Ml=Home_Consumption_Ml.drop(labels=['humidity '], axis=1)
```

```python
Home_Consumption_Ml['dewPoint']=Home_Consumption_Ml['dewPoint'].interpolate(method='linear')
```

```python
Home_Consumption_Ml
```

```python
Q1=Home_Consumption_Ml.quantile(0.25)
Q3=Home_Consumption_Ml.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
Home_Consumption_Ml.shape
```

```python
Home_Consumption_Ml_out=Home_Consumption_Ml[~((Home_Consumption_Ml<(Q1-1.5*IQR))|(Home_Consumption_Ml>(Q3+1.5*IQR))).any(axis=1)]
```

```python
Home_Consumption_Ml_out.shape
```

```python
PredictorsVariables=['gen [kW]','Dishwasher [kW]','Furnace 1 [kW]','Furnace 2 [kW]','Home office [kW]',
                     'Fridge [kW]', 'Wine cellar [kW]','Garage door [kW]','Kitchen 12 [kW]','Kitchen 14 [kW]',
                      'Kitchen 38 [kW]','Barn [kW]','Well [kW]','Microwave [kW]','Living room [kW]','Solar [kW]',
                      'temperature','humidity','pressure','windSpeed','dewPoint']

Targetvariable=['House overall [kW]']

```

```python
## Standardisation 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


P_Std=StandardScaler()
T_Std=StandardScaler()

X=P_Std.fit_transform(Home_Consumption_Ml_out[PredictorsVariables].values)
y=T_Std.fit_transform(Home_Consumption_Ml_out[Targetvariable].values)

```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=50)
```

```python
## Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn import metrics

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=50)

mlr=LinearRegression()
predictmodel=mlr.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
Predictedvar=T_Std.inverse_transform(predictions)
Y_origin=T_Std.inverse_transform(y_test)
Test_data=P_Std.inverse_transform(X_test)
print("Thr R2 Value",metrics.r2_score(y_train,predictmodel.predict(X_train)))
print('Accuracy on Test Data:',100- (np.mean(np.abs((Y_origin - Predictedvar) / Y_origin)) * 100))


```

```python
## Lasso Regrresion
from sklearn.linear_model import Lasso

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=50)

lass=Lasso(alpha=0.01)
predictmodel=lass.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
Predictedvar=T_Std.inverse_transform(predictions)
Y_origin=T_Std.inverse_transform(y_test)
Test_data=P_Std.inverse_transform(X_test)
print("Thr R2 Value",metrics.r2_score(y_train,predictmodel.predict(X_train)))
#print('Accuracy on Test Data:',100- (np.mean(np.abs((Y_origin - Predictedvar) / Y_origin)) * 100))
```
