```python
## Lasso Regression

from sklearn.linear_model import Lasso

Predictor_variable=['V_Component','WaterComponent','CoarseAggregateComponent','FineAggregateComponent','SuperplasticizerComponent']
Target_variable=['Strength']
X=ConcreteStrengthDataNew_Model[Predictor_variable].values
y=ConcreteStrengthDataNew_Model['Strength'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40,random_state=7890)

lass=Lasso(alpha=1e-02,max_iter=600,tol=0.0003)
predictmodel=lass.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
print("The R squared is",metrics.r2_score(y_train,predictmodel.predict(X_train)))
print(" The Accuracy score ",100- (np.mean(np.abs((y_test - predictions) / y_test)) * 100))



```

```python
## Ridge Reggresion

from sklearn.linear_model import Ridge

Predictor_variable=['V_Component','WaterComponent','CoarseAggregateComponent','FineAggregateComponent','SuperplasticizerComponent']
Target_variable=['Strength']
X=ConcreteStrengthDataNew_Model[Predictor_variable].values
y=ConcreteStrengthDataNew_Model['Strength'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.31,random_state=487901)
rig=Ridge(alpha=1e-02)
predictmodel=rig.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
print("The R squared is",metrics.r2_score(y_train,predictmodel.predict(X_train)))
print(" The Accuracy score ",100- (np.mean(np.abs((y_test - predictions) / y_test)) * 100))


```

```python
### Elastic Net regression

from sklearn.linear_model import ElasticNet

Predictor_variable=['V_Component','WaterComponent','CoarseAggregateComponent','FineAggregateComponent','SuperplasticizerComponent']
Target_variable=['Strength']
X=ConcreteStrengthDataNew_Model[Predictor_variable].values
y=ConcreteStrengthDataNew_Model['Strength'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.31, random_state=9250)


Enet=ElasticNet(alpha=0.01)
pedictmodel=Enet.fit(X_train,y_train)
predictions=pedictmodel.predict(X_test)
print("The R squared is",metrics.r2_score(y_train,predictmodel.predict(X_train)))
print(" The Accuracy score ",100- (np.mean(np.abs((y_test - predictions) / y_test)) * 100))



```

```python
## Decsion tree

from sklearn.tree import DecisionTreeRegressor

Predictor_variable=['V_Component','WaterComponent','CoarseAggregateComponent','FineAggregateComponent','SuperplasticizerComponent']
Target_variable=['Strength']
X=ConcreteStrengthDataNew_Model[Predictor_variable].values
y=ConcreteStrengthDataNew_Model['Strength'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.26, random_state=92)

dtree=DecisionTreeRegressor( max_depth=7,criterion='mse')
predictmodel=dtree.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
print("The R squared is",metrics.r2_score(y_train,predictmodel.predict(X_train)))
print(" The Accuracy score ",100- (np.mean(np.abs((y_test - predictions) / y_test)) * 100))

feautre_importance=pd.Series(predictmodel.feature_importances_,index=Predictor_variable)
feautre_importance.nlargest(10).plot(kind='bar')
```

```python
### Random foreest Regressor

from sklearn.ensemble import RandomForestRegressor


Predictor_variable=['V_Component','WaterComponent','CoarseAggregateComponent','FineAggregateComponent','SuperplasticizerComponent']
Target_variable=['Strength']
X=ConcreteStrengthDataNew_Model[Predictor_variable].values
y=ConcreteStrengthDataNew_Model['Strength'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.31, random_state=900)

rgf=RandomForestRegressor(n_estimators=200,criterion='mse',max_depth=5)
predictmodel=rgf.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
print("The R squared is",metrics.r2_score(y_train,predictmodel.predict(X_train)))
print(" The Accuracy score ",100- (np.mean(np.abs((y_test - predictions) / y_test)) * 100))

feautre_importance=pd.Series(predictmodel.feature_importances_,index=Predictor_variable)
feautre_importance.nlargest(10).plot(kind='bar')


```

```python
## Ada Boost Regressor

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor


Predictor_variable=['V_Component','WaterComponent','CoarseAggregateComponent','FineAggregateComponent','SuperplasticizerComponent']
Target_variable=['Strength']
X=ConcreteStrengthDataNew_Model[Predictor_variable].values
y=ConcreteStrengthDataNew_Model['Strength'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=91)

dtree=DecisionTreeRegressor(max_depth=5)
Ada =AdaBoostRegressor( base_estimator=dtree,n_estimators=300,learning_rate=0.002)
Predictmodel=Ada.fit(X_train,y_train)
predictions=Predictmodel.predict(X_test)
print("The R squared is",metrics.r2_score(y_train,predictmodel.predict(X_train)))
print(" The Accuracy score ",100- (np.mean(np.abs((y_test - predictions) / y_test)) * 100))



feautre_importance=pd.Series(predictmodel.feature_importances_,index=Predictor_variable)
feautre_importance.nlargest(10).plot(kind='bar')
```
